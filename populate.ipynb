```python
#import required packages
import mysql.connector
from mysql.connector import Error
import pandas as pd
```

```python
#create a connection to a mysql server
def create_server_connection(host_name, user_name, ):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")
    return connection

connection = create_server_connection("localhost", "root")
```

```python
# Create a database 
def create_database(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        print("Database created successfully")
    except Error as err:
        print(f"Error: '{err}'")

create_database_query = "CREATE DATABASE pollution_db"
create_database(connection, create_database_query)
```

```python
#connect to an exisiting database
def create_db_connection(host_name, user_name, db_name):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            database=db_name
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection
connection = create_db_connection("localhost", "root", "pollution_db")
```

```python
#read data from a csv file
data = pd.read_csv(r'C:\Users\Tripple A\Documents\ML Projects\MSC candidtate\DMF\clean.csv',low_memory=False)
len(data)
```

```python
data["Date Time"]=pd.to_datetime(data["Date Time"])
```

```python
# function to execute queries
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query successful")
    except Error as err:
        print(f"Error: '{err}'")
```

```python
# sql statement to create site table 
create_station_table='''CREATE TABLE IF NOT EXISTS pollution_db.`Stations` (
  `SiteID` INT NOT NULL,
  `Location` VARCHAR(45) NOT NULL,
  `geo_point_2d` VARCHAR(45) NOT NULL,
  PRIMARY KEY (`SiteID`),
  UNIQUE INDEX `SiteID_UNIQUE` (`SiteID` ASC) )
  ENGINE = InnoDB;'''
```

```python
# sql statement to create main reading table
create_readings_table='''CREATE TABLE IF NOT EXISTS pollution_db.`Readings` (
  `readingId` INT NOT NULL AUTO_INCREMENT,
  `Date Time` DATETIME NOT NULL,
  `NOx` FLOAT NULL,
  `NO2` FLOAT NULL,
  `NO` FLOAT NULL,
  `PM10` FLOAT NULL,
  `NVPM10` FLOAT NULL,
  `VPM10` FLOAT NULL,
  `NVPM2.5` FLOAT NULL,
  `PM2.5` FLOAT NULL,
  `VPM2.5` FLOAT NULL,
  `CO` FLOAT NULL,
  `O3` FLOAT NULL,
  `SO2` FLOAT NULL,
  `Temperature` FLOAT NULL,
  `RH` FLOAT NULL,
  `Air Pressure` FLOAT NULL,
  `SiteID` INT NOT NULL,
  `DateStart` DATETIME NULL,
  `DateEnd` DATETIME NULL,
  `InstrumentType` VARCHAR(45) NULL,
  `Current` VARCHAR(45) NULL,
  PRIMARY KEY (`readingId`),
  CONSTRAINT `fk_Readings_Station`
    FOREIGN KEY (`SiteID`)
    REFERENCES `pollution_db`.`Stations` (`SiteID`)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION)
ENGINE = InnoDB;
'''
```

```python
# excute all sql statement to create tables
print('Creating tables....')
execute_query(connection, create_station_table)
execute_query(connection, create_readings_table)
print("Tables are created....")
```

```python
# sql statement to populate stations table
try:
    cursor = connection.cursor()
    for i in data[["SiteID","Location", "geo_point_2d"]].drop_duplicates().values:  
        #here %S means string values 
        sql = "INSERT INTO pollution_db.Stations VALUES (%s,%s,%s)"
        cursor.execute(sql, tuple(i))
        print("Record inserted")
        # the connection is not auto committed by default, so we must commit to save our changes
        connection.commit()
except Error as e:
        print("Error while connecting to MySQL", e)
```
